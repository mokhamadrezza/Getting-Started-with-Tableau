# Distribusi Bernoulli

```python
# simulating 4 coin flips
import numpy as np
np.random.seed(42)
random_numbers = np.random.random(size=4)
random_numbers
```

```python
heads = random_numbers < 0.5
heads
```

```python
np.sum(heads)
```

```python
# simulating 4 coin flips
n_all_heads = 0 #initialize number of 4 head trials
for _ in range(10000):
        heads = np.random.random(size=4) < 0.5
        n_heads = np.sum(heads)
        if n_heads == 4:
           n_all_heads += 1
```

```python
n_all_heads / 10000
```

# Distribusi Uniform

# Distribusi Binomial

```python
# the binomial PMF
np.random.seed(42)
samples = np.random.binomial(60, 0.1, size=10000)
samples
```

```python
# Binomial PMF
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
n, p = 60, 0.1
x = [x for x in range(17)]

fig, ax = plt.subplots(1, 1)
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=5, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=3, alpha=0.5)
plt.xticks(x)
plt.ylabel('probability')
plt.xlabel('number of successes')
plt.title('Binomial PMF')
plt.show()
```

```python
# The Binomial CDF
import matplotlib.pyplot as plt
import seaborn as sns
# Generate binomial data
data_binom = binom.rvs(n=60,p=0.1,size=10000)
# Plot the CDF
sns.ecdfplot(data_binom)
plt.xlabel('number of successes')
plt.ylabel('CDF')
plt.title('Binomial CDF')
plt.show()
```

# Distribusi Poisson

```python
np.random.poisson(6, size=10000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
# Parameter distribusi Poisson
lambda_ = 6  # Rata-rata jumlah kejadian
# Nilai k (jumlah kejadian) dari 0 hingga 20
k_values = np.arange(0, 21)
# Menghitung PMF distribusi Poisson
poisson_pmf = poisson.pmf(k_values, lambda_)
# Plot PMF distribusi Poisson
plt.bar(k_values, poisson_pmf, color='blue', alpha=0.7, label=f'Poisson (λ={lambda_})')
plt.xlabel('Jumlah Kejadian (k)')
plt.ylabel('Probabilitas')
plt.title('Distribusi Poisson')
plt.xticks(k_values)
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Plot CDF distribusi Poisson
poisson_cdf = poisson.cdf(k_values, lambda_)
plt.step(k_values, poisson_cdf, color='blue', label=f'Poisson (λ={lambda_})')
plt.xlabel('Jumlah Kejadian (k)')
plt.ylabel('CDF')
plt.title('Distribusi Poisson')
plt.xticks(k_values)
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.show()
```

# Distribusi Normal

```python
# Generate Normal data
data_normal = np.random.normal(loc=0, scale=1, size=10000)
data_normal
# loc adalah nilai rata-rata dari distribusi normal
# scale adalah standar deviasi dari distribusi normal
```

```python
# Generate PMF Normal
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# Parameter distribusi Normal
mu = 0  # Rata-rata distribusi
sigma = 1  # Standar deviasi distribusi
# Nilai x dari -4 hingga 4
x = np.linspace(-4, 4, 1000)
# Menghitung PDF distribusi Normal
normal_pdf = norm.pdf(x, mu, sigma)
# Plot PDF distribusi Normal
plt.plot(x, normal_pdf, color='blue', label=f'Normal (μ={mu}, σ={sigma})')
plt.xlabel('Nilai (x)')
plt.ylabel('PDF')
plt.title('Distribusi Normal')
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Generate CDF Normal data
data_normal = np.random.normal(loc=0, scale=1, size=10000)
# Plot the CDF
sns.ecdfplot(data_normal)
plt.xlabel('x')
plt.ylabel('CDF')
plt.title('Normal CDF')
plt.show()
```

# Distribusi Eksponensial

```python
# Generate Exponential data with lambda = 0.5
data_exponential = np.random.exponential(scale=2, size=10000)
data_exponential
# scale adalah 1/lambda
# size adalah jumlah sampel yang dihasilkan oleh fungsi ini
```

```python
# Generate PDF Exponential
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
# Parameter distribusi Exponential
lambda_ = 0.5  # Parameter lambda
# Nilai x dari 0 hingga 10
x = np.linspace(0, 10, 1000)
# Menghitung PDF distribusi Exponential
exponential_pdf = expon.pdf(x, scale=1/lambda_)
# Plot PDF distribusi Exponential
plt.plot(x, exponential_pdf, color='blue', label=f'Exponential (λ={lambda_})')
plt.xlabel('Nilai (x)')
plt.ylabel('PDF')
plt.title('Distribusi Exponential')
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Generate CDF Exponential data
data_exponential = np.random.exponential(scale=2, size=10000)
# Plot the CDF
sns.ecdfplot(data_exponential)
plt.xlabel('x')
plt.ylabel('CDF')
plt.title('Exponential CDF')
plt.show()
```
